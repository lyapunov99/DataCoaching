```python
# This is a sample Python script.

# Press Shift+F10 to execute it or replace it with your code.
# Press Double Shift to search everywhere for classes, files, tool windows, actions, and settings.
from typing import List


class PPGObj:
    def __init__(self, name, sz, num_success):
        self.cohort_name = name  # gn = group or cohort name
        self.cohort_sz = sz      # gs = group or cohort size
        self.cohort_successes = num_success # num of successes for group
       # self.ppg = 0

    def get_gn(self):
        return self.cohort_name

    def set_gn(self, st):
        self.cohort_name = st

    def get_gs(self):
        return self.cohort_sz

    def set_gs(self, sz):
        self.cohort_sz = sz

    def get_s_count(self):
        return self.cohort_successes

    def set_gs_count(self, sc):
        self.cohort_successes = sc

    def get_ppg(self):
        return self.ppg

    def set_ppg(self, ppg):
        self.ppg = ppg

    def __repr__(self):
        return "Vector3( [ {0},{1},{2} ] )".format(self.cohort_name, self.cohort_sz, self.cohort_successes)

    def __str__(self):
        return "gn: {0}, gp_sz: {1},  gp_num_succ: {2}".format(self.cohort_name, self.cohort_sz, self.cohort_successes)


# Press the green button in the gutter to run the script.
if __name__ == '__main__':
    #  category_nm = ["AfAm", "Indig", "Asian", "Filipino", "Hispanic", "H/P Islander", "White", "Other"]
    #  cohort_sz = [63, 17, 112, 36, 761, 13, 1978, 202]
    #  num_successful_cohort = [29, 5, 77, 28, 357, 5, 1054, 96]
    # category_nm: list[str]
    category_nm, cohort_sz, num_successful_cohort = [], [], []
   #f = open("input-deanza.csv", 'r')
    
    f = open("C:/Users/khots/Google Drive/SWC/Data Coach/Model calculations and reports/Code/Python/input-deanza.csv", 'r')
    #f = open("C:/Users/khots/Google Drive/SWC/Data Coach/Model calculations and reports/Code/Python/input.csv", 'r')
    #    f = open("input.csv", 'r')
    next(f)  # skip the first line that contains column names

    # read information from the above file, and store into arraylists
    for line in f.readlines():
        fields = line.split(',')                        # e.g., returns ['african am', '5758', '3686\n']
        category_nm.append((fields[0]))                 # appends 'african am' to an array list
        cohort_sz.append(int(fields[1]))                # appends 5758 to an array list
        num_successful_cohort.append(int(fields[2]))    # appends 3686 to an array list
    f.close()

    # create a list. Each object contains 3 pieces of info: gpName, group count, group successes
    my_objects = []

    num_objects = len(category_nm)      # get total number of groups or cohorts

    for i in range(num_objects):        # create instances of objects, then append to creata list of objects
        my_objects.append(PPGObj(category_nm[i], cohort_sz[i], num_successful_cohort[i]))

    #for obj in my_objects:
     #   print(obj.cohort_name, obj.cohort_sz, obj.cohort_successes, sep=' ')

    cohort_total_size = sum(cohort_sz)  # determine sum total number of individuals
    print("Total number: ", cohort_total_size)
    cohort_total_successes = sum(num_successful_cohort)     # determine sum total number of successful individuals
    print("Total successful: ", cohort_total_successes)
    overall_success_rate = (cohort_total_successes / cohort_total_size)*100
    print("Overall Success Rate, all students  ", "{:>8.1f}".format(overall_success_rate), "%")

    print("\n", "       PPG minus one values:")
    for i in range(num_objects):
        categSuccessRate = my_objects[i].cohort_successes / my_objects[i].cohort_sz     # success rate for a cohort
        ppg_minus_one_success = (cohort_total_successes - my_objects[i].cohort_successes) / (cohort_total_size - my_objects[i].cohort_sz)
        ppgOut = "{:>8.1f}".format((categSuccessRate - ppg_minus_one_success) * 100)
        print("{:>18}".format(my_objects[i].cohort_name), ": ", ppgOut,"%")
```
